## Object Oriented Method

This method uses more code, but gives you more personalization, and control over graphics  
**pyplot** vs **OOM**  

Next the differences between these 2 are shown

* **plt.figure()** = Create a new figure, or activate an existing figure, the figure is the canvas where graphics will be displayed
* **plt.add_acces()** = Add an Axes to the figure, an axes is the defnied space in a canvas where a graphic will be displayed

### pyplot

### Object Oriented Method

```python
import numpy as np
import matplotlib.pyplot as ptl

#Creating the data
x = np.linspace(0,5,11)
y = x * x

#Creating the canvas
canv = plt.figure()
#Adding an axes, important: the parameters are the following, the first two are location in "x" and "y" plane, and the last 2 are size in "x" and "y" plane
ax1 = canv.add_axes([0.1,0.1,0.8,0.8])
#Adding the graph
ax1.plot(x,y)

```

```python
# Two graphs in one canvas
import numpy as np
import matplotlib.pyplot as ptl

#Creating the data
x = np.linspace(0,5,11)
y = x * x

#Creating the canvas
canv = plt.figure()
#Adding an axes
ax1 = canv.add_axes([0.1,0.1,0.8,0.8])
ax2 = canv.add_axes([0.17,0.55,0.3,0.3])
#Adding the graphs
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
# Two graphs in one canvas
import numpy as np
import matplotlib.pyplot as ptl

#Creating the data
x = np.linspace(0,5,11)
y = x * x

canv = plt.figure()
ax1 = canv.add_axes([0.1,0.1,0.8,0.8])
ax2 = canv.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y,"b")
ax2.plot(y,x,"g")
```

### Subplots in object oriented method

```python
import numpy as np
import matplotlib.pyplot as ptl

x = np.random.randint(1,10,11)
y = np.random.randint(1,10,11)
z = np.random.randint(1,10,11)
a = np.random.randint(1,10,11)

fig, axes = plt.subplots(nrows= 2, ncols=2)
axes[0,0].plot(x,y)
axes[1,0].plot(y,z)
axes[0,1].plot(z,a)
axes[1,1].plot(a,z)

```

```python
import numpy as np
import matplotlib.pyplot as ptl

x = np.random.randint(1,10,11)
y = np.random.randint(1,10,11)
z = np.random.randint(1,10,11)
a = np.random.randint(1,10,11)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2, ncols=2)
ax1.plot(x,y)
ax2.plot(y,z)
ax3.plot(z,a)
ax4.plot(a,z)
```

### Descriptions

* **axes.set_title** = Set a title for the Axes
* **axes.set_xlabel** = Set a description for axis x
* **axes.set_ylabel** = Set a description for axis y
* **fig, axes= (figsize=(0,0))** = Set the size of the canvas
* **axes.plot(label= "")** = Create a label to show the different aspects of the graph
* **axes.legend** = You need to run this function to activate the label function

```python
import numpy as np
import matplotlib.pyplot as ptl

x = np.random.randint(1,10,11)
y = np.random.randint(1,10,11)

fig, ((ax1,ax2)) = ptl.subplots(nrows= 1, ncols=2, figsize=(10,5))
ax1.plot(x,y,"g",label="XY")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_title("X versus Y")
ax1.legend()
ax2.plot(y,x,"b",label="YX")
ax2.set_title("Y versus X")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.legend()
```

### Colors and Styles

* **print(plt.style.available)** = Prints all the styles available for the canvas
* **ax.plot()**
    * **(alpha=x)** = Modifies the opacity of the line
    * **(linewidth=x)** = Modifies the width of the line
    * **(linestyle=x)** = Modifies the style of the line
    * **(marker=x)** = Add any marker at the points of data
    * **(markersize=x)** = Modifies the size of the markers
    * **(markerfacecolor=x)** = Modifies the color of the marker

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Create the data
x = [1,2,3,4,5]
x = np.array(x)
y = x**2

#Create canvas
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

#Add the linewidth,linestyle,marker,markersize,markerfacecolor functions
ax1.plot(x,y, linewidth=2,linestyle="--")
ax1.plot(x,x,linewidth=5,linestyle="dashed" )
ax1.set_title("linewidths and styles")

ax3.plot(x,y,marker="o",markersize=10,markerfacecolor="g")
ax3.set_title("markers, markers size and colors")

ax4.plot(x,y,marker="x",markersize=20,markerfacecolor="r")
ax4.set_title("markers, markers size and colors")


#Code obtain from: https://www.geeksforgeeks.org/change-the-line-opacity-in-matplotlib/#:~:text=Matplotlib%20allows%20you%20to%20regulate,such%20as%200.5%20or%200.25.
for i in range(10):
    ax2.plot([1, 2.8], [i]*2, linewidth=5, color='red', alpha=0.1*i)
    ax2.plot([3.1, 4.8], [i]*2, linewidth=5, color='green', alpha=0.1*i)
    ax2.plot([5.1, 6.8], [i]*2, linewidth=5, color='yellow', alpha=0.1*i)
    ax2.plot([7.1, 8.8], [i]*2, linewidth=5, color='blue', alpha=0.1*i)
      
for i in range(10):
    ax2.plot([1, 2.8], [-i]*2, linewidth=5, color='red', alpha=0.1*i)
    ax2.plot([3.1, 4.8], [-i]*2, linewidth=5, color='green', alpha=0.1*i)
    ax2.plot([5.1, 6.8], [-i]*2, linewidth=5, color='yellow', alpha=0.1*i)
    ax2.plot([7.1, 8.8], [-i]*2, linewidth=5, color='blue', alpha=0.1*i)

ax2.set_title("Alpha function, different levels of opacity")

```
