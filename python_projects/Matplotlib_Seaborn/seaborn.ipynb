# Seaborn  
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

## Introduction

### Import

```python
#Seaborn is imported in the next way, it is important to import matplotlib as well
import seaborn as sns
import matplotlib.pyplot as plt

```

### Basics

* **sns.set_theme()** = Set aspects of the visual theme for all matplotlib and seaborn plots.
* **sns.distplot()** = This function provides access to several approaches for visualizing the univariate or bivariate distribution of data.

```python
#Creating data
x = ["A","B","C"]
y = [1,2,3]

#Applying differente modes for set function
sns.barplot(x,y)
sns.set(style="dark", palette="muted", font_scale=1.2)
```

## Main Functions

### Categorical Values  
* **sns.countplot** = Show the counts of observations in each categorical bin using bars.
* **sns.stripplot** = Draw a scatterplot where one variable is categorical.
* **sns.swarmplot** = Draw a categorical scatterplot with non-overlapping points.
* **sns.boxplot** = Draw a box plot to show distributions with respect to categories.
* **sns.violinplot** = Draw a combination of boxplot and kernel density estimate.

### Countplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the countplot function
sns.countplot(data=tips, x="day", hue="sex")
```

### Stripplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the stripplot function
sns.stripplot(data=tips, x="day",y="total_bill",hue="sex", dodge=True)
```

### Swarmplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the swarmplot function
sns.swarmplot(data=tips, x="day",y="total_bill",hue="sex", dodge=True)
```

### Boxplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the boxplot function
plt.figure(figsize=(7,5))
sns.boxplot(data=tips, x="day",y="total_bill",hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day",y="total_bill",hue="sex", dodge=True)
```

### Violinplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the violinplot function
plt.figure(figsize=(7,5))
sns.violinplot(data=tips, x="day",y="total_bill",hue="sex", split=True)
```

## Relationships

* **sns.scatterplot** = Draw a scatter plot with possibility of several semantic groupings.
* **sns.lineplot** = Draw a line plot with possibility of several semantic groupings.
* **sns.relplot** = Figure-level interface for drawing relational plots onto a FacetGrid.

### Scatterplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the scatterplot function
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
tips.head(2)

```

### Linesplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the scatterplot function
sns.lineplot(data=tips, x="total_bill", y="tip", hue="sex")
```

### Relplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")

#Executing the scatterplot function
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex")
```

## Joinplots and Pairplots

* **sns.joinplots** = Draw a plot of two variables with bivariate and univariate graphs.
* **sns.pairplots** = Plot pairwise relationships in a dataset.

### Joinplot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")
tips.head(2)

#kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'

#Executing the joinplot function
sns.jointplot(data=tips,x="total_bill", y="tip", kind="kde", color="b")
```

### Pairplots

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")
tips.head(2)

#Executing the pairplot function
sns.pairplot(data=tips,hue="sex")
```

## Heatmap

* **sns.heatmap()** = Plot rectangular data as a color-encoded matrix.

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Loading data
tips = sns.load_dataset("tips")
#tips.head(5)

#Execute the heatmap function
sns.heatmap(tips.corr(), annot=True, linewidth=1,linecolor="white")
```
