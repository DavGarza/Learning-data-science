# Activation Functions for an artificial neuron

Is a function that is added into an artificial neural network in order to help the network learn complex patterns in the data.

Reference: https://towardsdatascience.com/everything-you-need-to-know-about-activation-functions-in-deep-learning-models-84ba9f82c253#:~:text=What%20is%20an%20activation%20function%3F,complex%20patterns%20in%20the%20data.

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Function

Used for maintaining values thorugh a process

```python
x = np.linspace(-10,10,num=1000)

def f(x):
    return x

y = f(x)

plt.plot(x,y)
```

## Heaviside Function  

Used for categorical values 

```python
x = np.linspace(-10,10,num=1000)

def f(x):
    return np.int0(x >= 0)

y = f(x)

plt.plot(x,y)
```

## Sigmoid Function  

$$f(x)=\frac{1}{1-e^{-x}}$$  

Used for Logistic Regression

```python
x = np.linspace(-10,10,num=1000)

def f(x):
    return 1/(1 + np.exp(-x))

y = f(x)

plt.plot(x,y) 
```

## Hiperbolic Tangent Function

*Function already in numpy*


```python
x = np.linspace(-10,10,num=1000)

def f(x):
    return np.tanh(x)

y = f(x)

plt.plot(x,y) 
```

## ReLU Function  

```python
x = np.linspace(-10,10,num=1000)

def f(x):
    return np.maximum(x,0)

y = f(x)

plt.plot(x,y) 
```
