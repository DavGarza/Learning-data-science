# Mathematical Functions

Definition: a relationship or expression involving one or more variables.

## Lineal Function

Formulas
* Form of lineal function: *f(x) = mx + b*
* Formula to find m: *m = (y2-y1)/(x2-x1)



```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Establish the constants, m control the angle of the line, and the b control where the functions starts
m = 1
b = 0

#Create a function
def f(x):
    return m*x + b 

#Create the data points for the variable f(x)
x = np.linspace(-50,50,100)

#Apply the formula where the variable y gets trough f(x)
y = f(x)

#Visualize the function
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

## Polynomic Function

Definition: Is an expression consisting of indeterminates (also called variables) and coefficients, that involves only the operations of addition, subtraction, multiplication, and non-negative integer exponentiation of variables.  

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Establish the constants, m control the angle of the line, and the b control where the functions starts
m = 1
b = 0

#Create functions

#second degree function, all the pair degree functions look similar, see the chart below
def f(x):
    return (3*x**2) + (4*x)

#third degree function, all odd degree function look similar, see the chart below
def g(x):
    return (5*x**3) + (2*x**2) - (3*x) + 7

#Create the data points for the variable f(x)
x = np.linspace(-10,10,100)

#Apply the formula where the variable y gets trough f(x)
y = f(x)
y2 = g(x)

fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.plot(x,y)
ax1.set_title("Pair degree function")
ax2.plot(x,y2)
ax2.set_title("Odd degree function")
```

### Trigonometric Function

Definition: Are real functions which relate an angle of a right-angled triangle to ratios of two side lengths.

And have the next form

$$f\left(x\right)=a^x$$
where $a\:>0\:and\:a\:\ne 1$

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Create the data points for the variable x
x = np.linspace(-10,10,1000)



#Define the functions and add the y variable
def f(x):
    return np.sin(x)

def g(x):
    return np.cos(x)

def l(x):
    return np.tan(x)

def m(x):
    return np.arcsin(x)

y = f(x)
y2 = g(x)
y3 = l(x)
y4 = m(x)

#Create the graphics

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2, ncols=2, figsize=(10,7))
ax1.plot(x,y)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
ax1.set_title("Sine of x")
ax2.set_title("Cosine of x")
ax3.set_title("Tangent of x")
ax4.set_title("Cosecant of x")
```

### Exponential Function



$$f\left(x\right)=a^x$$
where $a\:>0\:and\:a\:\ne 1$

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Create the data points for the variable x
x = np.linspace(2,5,1000)

#Define the functions and add the y variables
def f(x):
    return (x**2)

def g(x):
    return (x**3)

def s(x):
    return (x**4)

def l(x):
    return (x**5)

y = f(x)
y2 = g(x)
y3 = s(x)
y4 = l(x)

plt.plot(x,y, label="y= x Squared 2")
plt.plot(x,y2, label="y= x Squared 3")
plt.plot(x,y3, label="y= x Squared  4")
plt.plot(x,y4, label="y= x Squared  5")

plt.legend()
```

## Function Composition

Definition: Is an operation that takes two functions, and produces a new function.  

$$
f ∘ g = (f ∘ g)(x) = f(g(x))
$$


```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10, num=1000)

def g(x):
    return (5*x**3)

def f(x):
    return np.sin(x)

y = f(g(x))

plt.plot(x,y)
```

## Function Manipulation

Formulas for manipulating the function


**Moving functions**

$$
c > 0
$$

- $y=f(x)+c$ moves $c$ units up.
- $y=f(x)-c$ moves $c$ units down.
- $y=f(x-c)$ moves $c$ units right.
- $y=f(x+c)$ moves $c$ units left.


```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10, num=1000)

def f(x):
    return (x**2)

#Moving up
y = f(x) + 25
#Moving down
y2 = f(x) - 25
#Moving right
y3 = f(x-2)
#Moving left
y4 = f(x+2)


fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2, ncols=2, figsize=(10,7))
ax1.plot(x,y)
ax1.set_title("Move up")
ax1.axhline(y=0, color='r')
ax1.axvline(x=0, color='r')
ax2.plot(x,y2)
ax2.set_title("Move down")
ax2.axhline(y=0, color='r')
ax2.axvline(x=0, color='r')
ax3.plot(x,y3)
ax3.set_title("Move right")
ax3.axhline(y=0, color='r')
ax3.axvline(x=0, color='r')
ax4.plot(x,y4)
ax4.set_title("Move left")
ax4.axhline(y=0, color='r')
ax4.axvline(x=0, color='r')
```

**Elongation and Compresion**

$$
c > 0
$$

- $y=c \cdot f(x)$ elongate the graph verticaly in a factor of $c$.
- $y= \frac{1}{c} \cdot f(x)$ compress the grph verticaly in a factor of $c$.
- $y=f(c \cdot x)$ compress the graph horizontaly in a factor of $c$.
- $y= f(\frac{1}{c} \cdot x )$ elongate the graph in a factor of $c$.

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Create the data points for the variable x
x = np.linspace(-15,15,1000)
c = 2
#Define the functions and add the y variable
def f(x):
    return np.sin(x)

#Elongate vert
y = c*f(x)
#Compress vert
y2 = (1/c)*f(x) - 25
#Compress hor
y3 = f(c*x)
#Elongate hor
y4 = f((1/c)*x)


fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2, ncols=2, figsize=(10,7))
ax1.plot(x,y)
ax1.set_title("Elongate vert")
ax1.axhline(y=0, color='r')
ax1.axvline(x=0, color='r')
ax2.plot(x,y2)
ax2.set_title("Compress vert")
ax2.axhline(y=0, color='r')
ax2.axvline(x=0, color='r')
ax3.plot(x,y3)
ax3.set_title("#Compress hor")
ax3.axhline(y=0, color='r')
ax3.axvline(x=0, color='r')
ax4.plot(x,y4)
ax4.set_title("Elongate hor")
ax4.axhline(y=0, color='r')
ax4.axvline(x=0, color='r')
```
