## II. Distribution of Probability

### II.I Continious Probability Distribution 

A probability distribution is a statistical function that describes all the possible values and likelihoods that a random variable can take within a given range.

Reference : https://www.investopedia.com/terms/p/probabilitydistribution.asp#:~:text=A%20probability%20distribution%20is%20a,
take%20within%20a%20given%20range.   

***P(x ≥ 0)***

Given P, what is the probability of x ≥ 0


```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
  
# A custom function to calculate
# probability distribution function
def pdf(x):
    mean = np.mean(x)
    std = np.std(x)
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
    return y_out
  
# To generate an array of x-values
x = np.arange(-2, 2, 0.1)
  
# To generate an array of 
# y-values using corresponding x-values
y = pdf(x)
  
# To fill in values under the bell-curve
x_fill = np.arange(-2, 2, 0.1)
y_fill = pdf(x_fill)
  
# Plotting the bell-shaped curve
plt.style.use('seaborn')
plt.figure(figsize = (6, 6))
plt.plot(x, y, color = 'black',
         linestyle = 'dashed')
  
plt.scatter(x, y, marker = 'o',
            s = 25, color = 'red')
  
plt.fill_between(x_fill, y_fill, 0,
                 alpha = 0.3, color = 'blue', where= x >= 0)
plt.show()
```

### II.II Discrete Probability Distribution or Bernoulli(Binomial) Probability Distribution

A frequency distribution of the possible number of successful outcomes in a given number of trials in each of which there is the same boolean discrete probability of success.

Given *n* as the number of repetitions and *p* the probability, and *k* the exact succeses in *n* trials

$$
  \ (n k) \, = \frac{\ n!}{k!(n-k)!}
$$

Consider the next situation

There is 3 coins, and the coins are thrown   
What is the probability of getting 2 heads

$$
  \ (3,2) \, = \frac{\ 3}{2!(3-2)!}
$$




The formula above considers a fair probability between the 2 possible values, but sometimes it isn´t, so the formula for when it insn´t fair is   

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$   

And we can make transform it into a pyhton function like the next one

```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt

def binomial(k, n, p):
  return factorial(n)/(factorial(k)*(factorial(n-k)))*pow(p,k)*pow(1-p, n-k)

## We can try the function with the same example above
binomial(2,3, 0.5)
```
